## Remove Outliers

```python
def remove_outliers(df, columns):
    for col in columns:
        # Identify outliers
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        # Mask outliers
        outlier_mask = (df[col] < lower_bound) | (df[col] > upper_bound)

        # Remove outliers
        df.drop(df[outlier_mask].index, inplace=True)

# Specify columns to remove outliers from
columns_to_clean = ['Hours', 'streams','Age']


remove_outliers(df, columns_to_clean)
df_filtered = df
print(df_filtered)
```

# Model Implementation

## 1. MLP Classifier

```python
# Import Library
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score, f1_score, confusion_matrix


# Initialize the MLP Classifier
mlp_classifier = MLPClassifier(random_state=42)

# Train the MLP Classifier
mlp_classifier.fit(x_train, y_train)
```

```python
# Setting up the class labels
class_names = ['None', 'Mild', 'Moderate', 'High', 'Extreme']
# Predictions and evaluation for MLP CLassifier
mlp_predictions = mlp_classifier.predict(x_test)
```

```python
# Precision, Recall, F1-score
mlp_accuracy = accuracy_score(y_test, mlp_predictions)
mlp_precision = precision_score(y_test, mlp_predictions, average='weighted')
mlp_recall = recall_score(y_test, mlp_predictions, average='weighted')
mlp_f1 = f1_score(y_test, mlp_predictions, average='weighted')
print("MLP Classifier Accuracy:", mlp_accuracy)
print(f"Precision: {mlp_precision}")
print(f"Recall: {mlp_recall}")
print(f"F1-score: {mlp_f1}")
```

```python
def take_fixed_options_mcq_quiz(questions):
    # Define options for different questions
    general_options = ["Not At All", "A Little Bit", "Somewhat", "Very Much", "Extremely"]
    numeric_options = ["1", "2", "3", "4", "5"]
    age_options = ["18-25", "26-35", "36-45", "46-55", "56+"]
    narcissism_options = ["Very Low", "Low", "Moderate", "High", "Very High"]
    job_options = ["Unemployed", "Part-time", "Full-time", "Student", "Retired"]
    degree_options = ["None", "High School", "Associate", "Bachelor's", "Master's or Higher"]

    # Map questions to their respective options
    options_map = {
        1: general_options, 2: general_options, 3: general_options, 4: general_options,
        5: general_options, 6: general_options, 7: general_options, 8: general_options,
        9: general_options, 10: general_options, 11: general_options, 12: general_options,
        13: general_options, 14: general_options, 15: general_options, 16: general_options,
        17: general_options, 18: numeric_options, 19: numeric_options, 20: age_options,
        21: narcissism_options, 22: job_options, 23: degree_options
    }

    # Initialize a list to store user responses
    user_responses = []

    # Display questions and get user responses
    for i, question in enumerate(questions, start=1):
        print(f"\nQuestion {i}: {question}")
        options = options_map[i]
        for j, option in enumerate(options, start=1):
            print(f"{j}. {option}")

        # Get user input for the selected option with exception handling
        while True:
            try:
                user_input = int(input("Your answer (enter the option number): "))

                # Validate user input
                if 1 <= user_input <= len(options):
                    break
                else:
                    print("Invalid input. Please enter a valid option number.")
            except ValueError:
                print("Invalid input. Please enter a valid integer.")

        # Store the user response
        user_responses.append(user_input)

    # Display user responses
    print("\nYour selected answers:")
    for i, response in enumerate(user_responses, start=1):
        print(f"Question {i}: {options_map[i][response - 1]}")

    # Process responses for prediction
    dd = []
    dfs1 = user_responses[:6]
    dfs2 = user_responses[6:13]
    dfs3 = user_responses[13:17]
    dfs4 = user_responses[17:18]
    dfs5 = user_responses[18:19]
    dfs6 = user_responses[19:20]
    dfs7 = user_responses[20:21]
    dfs8 = user_responses[21:22]
    dfs9 = user_responses[22:23]

    d1 = np.array(dfs1)
    d2 = np.array(dfs2)
    d3 = np.array(dfs3)
    d4 = np.array(dfs4)
    d5 = np.array(dfs5)
    d6 = np.array(dfs6)
    d7 = np.array(dfs7)
    d8 = np.array(dfs8)
    d9 = np.array(dfs9)

    # Calculate medians
    dd.append(np.median(d1).round().astype('int64'))
    dd.append(np.median(d2).round().astype('int64'))
    dd.append(np.median(d3).round().astype('int64'))
    dd.append(np.median(d4).round().astype('int64'))
    dd.append(np.median(d5).round().astype('int64'))
    dd.append(np.median(d6).round().astype('int64'))
    dd.append(np.median(d7).round().astype('int64'))
    dd.append(np.median(d8).round().astype('int64'))
    dd.append(np.median(d9).round().astype('int64'))

    # Create the numpy array
    symptom = np.array(["none", "Mild", "Moderate", "High", "Extreme"])

    df_val = pd.DataFrame([dd], columns=[
        'Fear_Component',
        'Avoidance_Component',
        'Physiological_Discomfort_Component',
        'Hours',
        'streams',
        'Age',
        'Narcissism',
        'Work',
        'Degree'
    ])

    # Make predictions on data
    df_predictions = mlp_classifier.predict(df_val)
    print("Social Phobia Prediction is: ", symptom[df_predictions][0])

# Example questions (You can customize these questions)
quiz_questions = [
    "I am afraid of people in authority.",
    "I am bothered by blushing in front of people.",
    "Parties and social events scare me.",
    "I avoid talking to people I donâ€™t know.",
    "Being criticized scares me a lot.",
    "I avoid doing things or speaking to people for fear of embarrassment.",
    "Sweating in front of people causes me distress.",
    "I avoid going to parties.",
    "I avoid activities in which I am the center of attention.",
    "Talking to strangers scares me.",
    "I avoid having to give speeches.",
    "I would do anything to avoid being criticized.",
    "Heart palpitations bother me when I am around people.",
    "I am afraid of doing things when people might be watching.",
    "Being embarrassed or looking stupid are among my worst fears.",
    "I avoid speaking to anyone in authority.",
    "Trembling or shaking in front of others is distressing to me.",
    "I spend hours for video games.",
    "I use streams for games.",
    "I am years old.",
    "My Narcissism in games.",
    "My job nature.",
    "My Degree is.",
]

# Call the function with the example questions
take_fixed_options_mcq_quiz(quiz_questions)
```
